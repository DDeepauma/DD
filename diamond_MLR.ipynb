Data Cleaning & Encoding:

```python
df2 = df.drop(columns = ['Depth %','Table %','Length','Width','Height','Cut','Polish','Symmetry','Girdle','Culet','Type','Fluorescence'])
```

```python
df2.info()
```

```python
df2
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df2['Shape'] = le.fit_transform(df2['Shape'])
```

```python
df2['Color'] = le.fit_transform(df2['Color'])
```

```python
df2['Clarity'] = le.fit_transform(df2['Clarity'])
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
df2=df2.dropna()
```

```python
df2.info()
```

X,Y Split

```python
x = df2.iloc[:,0:5]
y = df2.iloc[:,5]
```

```python
x
```

```python
y
```

Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=345)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Model Building: MLR

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
model1.fit(x_train,y_train)
```

Model Evaluation:

```python
y_predict1 = model1.predict(x_test)
```

```python
y_predict1
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
MAE1 = mean_absolute_error(y_test,y_predict1)
MSE1 = mean_squared_error(y_test,y_predict1)
RMSE1 = np.sqrt(MSE1)
R2_1 = r2_score(y_test,y_predict1)
```

```python
print(model1.coef_)
```

Model Building: Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model2 = MultinomialNB()
```

```python
model2.fit(x_train,y_train)
```

Model Evaluation:

```python
y_predict2 = model2.predict(x_test)
```

```python
MAE2 = mean_absolute_error(y_test,y_predict2)
MSE2 = mean_squared_error(y_test,y_predict2)
RMSE2 = np.sqrt(MSE2)
R2_2 = r2_score(y_test,y_predict2)
```

Model Building: SVR

```python
from sklearn.svm import SVR
```

```python
model3 = SVR()
```

```python
model3.fit(x_train, y_train)
```

Model Evaluation:

```python
y_predict3 = model3.predict(x_test)
```

```python
y_predict3
```

```python
MAE3 = mean_absolute_error(y_test,y_predict3)
MSE3 = mean_squared_error(y_test,y_predict3)
RMSE3 = np.sqrt(MSE3)
R2_3 = r2_score(y_test,y_predict3)
```

MAE, MSE, R2 value of all 3 models:

```python
print (f"Multinomial Naive Bayes model MAE value   : ","{:.2}".format(MAE2))
print (f"Support vector machine regressor MAE value: ","{:.2}".format(MAE3))
print (f"Multilinear regression MAE value          : ","{:.2}".format(MAE1))
print (f"____________________________________________________________")
print (f"Multinomial Naive Bayes model MSE value   : ","{:.2}".format(MSE2))
print (f"Support vector machine regressor MSE value: ","{:.2}".format(MSE3))
print (f"Multilinear regression MSE value          : ","{:.2}".format(MSE1))
print (f"____________________________________________________________")
print (f"Multinomial Naive Bayes model R2 value    : ","{:.2}".format(R2_2))
print (f"Support vector machine regressor R2 value : ","{:.2}".format(R2_3))
print (f"Multilinear regression R2 value           : ","{:.2}".format(R2_1))
```
