Data Cleaning & Encoding:

```python
df1 = df.copy()
```

```python
df1=df1.drop(['PassengerId','Name', 'Ticket', 'Cabin'], axis = 1)
```

```python
df1.isnull().sum()
```

```python
df1= df1.dropna()
```

```python
df1.isnull().sum()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df1['Sex'] = le.fit_transform(df1['Sex'])
```

```python
df1['Embarked'] = le.fit_transform(df1['Embarked'])
```

```python
df1
```

X,Y Split

```python
x = df1.iloc[:,[3,6]].values
```

```python
x
```

```python
y = df1.iloc[:,0].values
```

```python
y
```

Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Model Building: GaussianNB, BernoulliNB

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB
```

```python
model1 = GaussianNB()
model2 = BernoulliNB()
```

```python
model1.fit(x_train,y_train)
```

```python
model2.fit(x_train,y_train)
```

Model Evaluation: GaussianNB, BernoulliNB

```python
y_predict1 = model1.predict(x_test)
y_predict2 = model2.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm1 = confusion_matrix(y_test,y_predict1)
cm2 = confusion_matrix(y_test,y_predict2)
```

```python
cm1
```

```python
cm2
```

Model Building: SVC

```python
from sklearn.svm import SVC
```

```python
model3 = SVC()
```

```python
model3.fit(x_train,y_train)
```

Model Evaluation: SVC

```python
y_predict3 = model3.predict(x_test)
```

```python
cm3 = confusion_matrix(y_test,y_predict3)
```

```python
cm3
```

Model Building: LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lor = LogisticRegression(random_state=0, solver='lbfgs')
```

```python
lor.fit(x_train, y_train)
```

Model Evaluation: LogisticRegression

```python
y_predictlor = lor.predict(x_test)
```

```python
cm4 = confusion_matrix(y_test,y_predictlor)
```

```python
cm4
```

Data visualization of 4 models:

```python
sns.set(font_scale=2)
ndarray:cm1
ndarray:cm2
ndarray:cm3
ndarray:cm4

label = [1,2]

# Create a figure with three subplots
fig, axs = plt.subplots(1,4, figsize=(20,6))

fig.suptitle('Titanic Survey - Confusion Matrix')

# Plot the first confusion matrix
s = sns.heatmap(cm1, ax=axs[0], annot=True, fmt='d', cmap='crest',linewidths=.5)
axs[0].set_title('Gaussian Naive Bayes')
axs[0].set_xlabel('Predicted Labels')
axs[0].set_ylabel('True Labels')
axs[0].set_xticks([0, 1])
axs[0].set_yticks([0, 1])
axs[0].set_xticklabels(label)
axs[0].set_yticklabels(label)

# Plot the second confusion matrix
sns.heatmap(cm2, ax=axs[1], annot=True, fmt='d', cmap='icefire',linewidths=.5)
axs[1].set_title('Bernoulli Naive Bayes')
axs[1].set_xlabel('Predicted Labels')
axs[1].set_ylabel('True Labels')
axs[1].set_xticks([0, 1])
axs[1].set_yticks([0, 1])
axs[1].set_xticklabels(label)
axs[1].set_yticklabels(label)

# Plot the third confusion matrix
sns.heatmap(cm3, ax=axs[2], annot=True, fmt='d', cmap = 'rocket',linewidths=.5)
axs[2].set_title('SVM')
axs[2].set_xlabel('Predicted Labels')
axs[2].set_ylabel('True Labels')
axs[2].set_xticks([0, 1])
axs[2].set_yticks([0, 1])
axs[2].set_xticklabels(label)
axs[2].set_yticklabels(label)

# Plot the fourth confusion matrix
sns.heatmap(cm4, ax=axs[3], annot=True, fmt='d', cmap = 'flare',linewidths=.5)
axs[3].set_title('LoR')
axs[3].set_xlabel('Predicted Labels')
axs[3].set_ylabel('True Labels')
axs[3].set_xticks([0, 1])
axs[3].set_yticks([0, 1])
axs[3].set_xticklabels(label)
axs[3].set_yticklabels(label)

# Layout so plots do not overlap
fig.tight_layout()

plt.show()
```

Data accuracy of 4 models:

```python
from sklearn.metrics import accuracy_score
acc1 = accuracy_score(y_test,y_predict1)
acc2 = accuracy_score(y_test,y_predict2)
acc3 = accuracy_score(y_test,y_predict3)
acc4 = accuracy_score(y_test,y_predictlor)
print (f"Gaussian Naive Bayes model accuracy : ","{:.2%}".format(acc1))
print (f"Bernoulli Naive Bayes model accuracy: ","{:.2%}".format(acc2))
print (f"SVM model accuracy                  : ","{:.2%}".format(acc3))
print (f"Logistic Regression model accuracy  : ","{:.2%}".format(acc4))
```

The result based on the above four Confusion Matrix of Accuracy:
SVM model accuracy is best
